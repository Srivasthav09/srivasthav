```python
import pandas as pd
import matplotlib.pyplot as plt
def data(path):
  """
  read data using pandas 
  function use read_csv 
  path pass as perameter
  """
  datad=pd.read_csv(path)
  return datad
```

```python
df=data("/content/population.csv")
df.head(5)
df.replace('..', 0, inplace=True)
col=df.columns[4:]

def py_plot_Series(col):
  """
  create box line and bar plot
  """
  df[col] = df[col].astype(float)
  df.plot(x='Series Name', y=col, kind='line')
  plt.xlabel('Series Name')
  plt.ylabel(col)
  plt.title('Value by Series Name')
  plt.show()
  # box plot
  df[col] = df[col].astype(float)
  df.plot(x='Series Name', y=col, kind='box')
  plt.xlabel('Series Name')
  plt.ylabel(col)
  plt.title('Value by Series Name')
  plt.show()
  # bar plot
  df[col] = df[col].astype(float)
  df.plot(x='Series Name', y=col, kind='bar')
  plt.xlabel('Series Name')
  plt.ylabel(col)
  plt.title('Value by Series Name')
  plt.show()

  df[col] = df[col].astype(float)
  df.plot(x='Series Name', y=col)
  plt.xlabel('Series Name')
  plt.ylabel(col)
  plt.title('Value by Series Name')
  plt.show()
for i in col:
  py_plot(i)
```
